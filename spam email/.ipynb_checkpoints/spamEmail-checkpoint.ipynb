```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
train_df=pd.read_csv('train_data.csv')
train_df
```

```python
train_df.isnull().sum()
```

```python
test_df=pd.read_csv('test_features.csv')
test_df
```

```python
test_df.isnull().sum()
```

```python
train_df=train_df.drop('Id',axis=1)
test_df=test_df.drop('Id',axis=1)
```

```python
train_df.head()
```

```python
train_df['ham'].replace(to_replace=[True,False], value=[1,0], inplace=True)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
X=np.asanyarray(train_df.drop('ham',axis=1))
X[0:2]
```

```python
y=np.asanyarray(train_df['ham'])
y[0:2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5000)
```

```python
print(f'Train Dataset X and Y : {X_train.shape}, {y_train.shape}')
print(f'Test Dataset X and Y :  {X_test.shape}, {y_test.shape}')
```

```python
from naive_bayes import NaiveBayes 
```

```python
def accuracy(y_true, y_pred):
    # sourcery skip: inline-immediately-returned-variable
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy
```

```python
nb = NaiveBayes()
nb.fit(X_train, y_train)
predictions = nb.predict(X_test)
print(X_test[0])
print(f'Naive Bayes classification accuracy {accuracy(y_test, predictions)}')
```

```python
test_data=np.asanyarray(test_df)
predictions2=nb.predict(test_data)
print(test_data[0])
print(f'Naive Bayes classification accuracy {accuracy(y_test, predictions2)}')
```
